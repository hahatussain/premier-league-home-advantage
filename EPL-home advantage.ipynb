```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (20, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
df_by_season =  df.groupby(['Season']).sum()
df_by_season
```

```python
df_by_season['Total Goals'] = df_by_season['FTHG'] + df_by_season['FTAG']
df_by_season = df_by_season.reset_index()
df_by_season
```

```python
df_melted = df_by_season[['Season', 'FTHG', 'FTAG']]
df_melt = df_melted.melt("Season", var_name="H/A", value_name="Goals")
df_melt
```

```python
sns.barplot('Goals', 'Season', hue = 'H/A', data = df_melt)
plt.title('Number of Goals Scored Per Season') 
plt.ylabel('Season')
plt.legend(['Home Goals', 'Away Goals'])
```

```python
df_by_season['FTHG-FTAG']=df_by_season['FTHG']-df_by_season['FTAG']
df_by_season
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.lineplot(data = df_by_season, x="Season", y="FTHG-FTAG")
```
